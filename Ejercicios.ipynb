```python
#cuantos elementos hay en la lista
lst = [1, 2, 34, 5, 3, 12, 9, 8, 67, 89, 98, 90, 39, 21, 45, 46, 23, 13]

cantidad = len(lst)

print("La cantidad de elementos en la lista es:", cantidad)
```

```python
#cual es el primer elemento en la lista 
x = [1, 2, 34, 5, 3, 12, 9, 8, 67, 89, 98, 90, 39, 21, 45, 46, 23, 13]

primer_elemento = x[0]
print(primer_elemento)
```

```python
#el último elemento en la lista 
ultimo = x[len(x) - 1]
print(ultimo)
```

```python
#cuál es el indice del elemento 90 en la lista
x = [1, 2, 34, 5, 3, 12, 9, 8, 67, 89, 98, 90, 39, 21, 45, 46, 23, 13]

indice = x.index(90)
print(indice)
```

```python
#escribe un programa que busque un elemento especifico en una lista dada, si el elemento no se enceuntra, imprime un mensaje indicando que no se encuentra en la lista 

names = ["Alice", "Bob", "Charlie", "Dave"]

search_name = input("Ingresa el nombre a buscar: ")

if search_name in names:
    print("El nombre se encuentra en la lista")
else:
    print("No se encuentra en la lista")

```
