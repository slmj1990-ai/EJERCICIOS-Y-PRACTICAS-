```python
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/data-bootcamp-v4/prework_data/main/students_performance.csv"
```

```python
df = pd.read_csv(url)
```

```python
print(df.head())
```

```python
print(df.tail())
```

```python
print(df.info())
```

```python
print(df.describe())
```

```python
print(df.dtypes)
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
plt.hist(df["math score"], bins=10)
plt.title("Histograma de Math Score")
plt.xlabel("Puntaje")
plt.ylabel("Frecuencia")
plt.show()
```

```python
sns.barplot(x="gender", y="math score", data=df)
plt.title("Promedio de Math Score por Gnero")
plt.show()
```

```python
conteo_genero = df["gender"].value_counts()
conteo_genero

```

```python
plt.pie(
    conteo_genero,
    labels=conteo_genero.index,
    autopct="%1.1f%%",
    startangle=90
)

plt.title("Distribucin de estudiantes por gnero")
plt.show()
```
